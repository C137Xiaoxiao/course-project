```python
def nnet(X):
  out = hk.Linear(20)(X)
  out = jax.nn.relu(out)

  out = hk.Linear(20)(X)
  out = jax.nn.relu(out)

  out = hk.Linear(3)(out)
  out = jax.nn.softmax(out)
  return out
```

```python
init, nnet = hk.without_apply_rng(hk.transform(nnet))
```

```python
optimizer = optax.adam
lr = 1e-3
rng = jax.random.PRNGKey(0)
Θ = init(rng, X[0])
opt_state = optimizer(lr).init(Θ)
```

```python
def cross_entropy(prob, y):
  from jax.scipy.special import xlogy
  return -xlogy(y, prob).sum()
  # return - (y * jnp.log(prob)).sum()


def L(Θ):
  prob = nnet(Θ, X)
  return jnp.mean(cross_entropy(prob, y))


@jax.jit
def update(Θ, opt_state):
    def L(Θ):
      predictions  = nnet(Θ, X)
      return cross_entropy(predictions, y)

    grads = jax.grad(L)(Θ)
    updates, opt_state = optimizer(lr).update(grads, opt_state)
    Θ = optax.apply_updates(Θ, updates)
    return Θ,  opt_state



```

```python
for iteration in range(100000):
  Θ, opt_state = update(Θ, opt_state)
```

```python
prediction = jnp.argmax(nnet(Θ, X), 1)
```
